````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example dataset, house size in sq ft and corresponding prices
x = np.array([[500], [700], [900], [1100], [1300]])
y = np.array([150000, 200000, 250000, 300000, 350000])

# Create and train the model
model = LinearRegression()

model.fit(x,y)

# Predict house prices for a new house of size 1000 sq ft

new_size = np.array([[1000]])
predicted_price = model.predict(new_size)

print(f"Predicted price for a 1000 sq ft house is $ {predicted_price[0]:,.2f}")
````

````python
plt.scatter(x,y, color='blue', label='Actual Prices')

plt.plot(x, model.predict(x), color='red', label='Regression Line')

plt.scatter(new_size, predicted_price, color='green', label='Prediction')

plt.xlabel("House Size (sq ft)")
plt.ylabel("Price ($)")

plt.legend()
plt.show()
````
