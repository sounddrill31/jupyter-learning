## Create a simple plot using numpy, pandas dataframe, matplotlib

### 1. Take your monthly expenditure on canteen. Example, daily you may spend like 50 or 80 rupees etc.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Input data
X = np.array([[45],[87],[42],[100],[48],[80], [70], [50], [30], [46]])
dates = ["05-03-2025","06-03-2025","07-03-2025","08-03-2025","09-03-2025","10-03-2025", "11-03-2025", "12-03-2025", "13-03-2025", "14-03-2025"]

# Create a line graph
plt.figure(figsize=(10, 6))
plt.plot(dates, X.flatten(), marker='o', color='blue', linestyle='-', alpha=0.7)
plt.title('Daily Expenditure')
plt.xlabel('Date')
plt.ylabel('Expenditure')
plt.xticks(rotation=45)
plt.grid()
plt.legend(["In Rupees"])

# Show the plot
plt.tight_layout()
plt.show()
```

### 2. Using Matplotlib, create a Pie Chart.

```python
import matplotlib.pyplot as plt
import numpy as np 
slices = [7,2,2,13,23]
acts = ["sleep", "eat", "work", "play","music"] 
cols = ['c', 'm', 'r', 'b', 'g'] 

print("Words To Choose:")
for i in range(len(acts)):
    print(f"{i}. {acts[i]}")

pullout = acts.index(input(f"Enter The Above Word To Explode: "))
tuplex = np.zeros(len(acts))
tuplex[pullout] = 0.5
plt.pie(slices, labels = acts, colors= cols,
    startangle = 90,
    shadow=True,
    explode = (tuplex),
    autopct='%1.1f%%')

plt.title('Pie chart')
plt.show()
```
